```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns

np.random.seed(100)
x = np.random.sample(10)
y = 0.4*x + 2
df = pd.DataFrame({'x':x,'y':y})

plt.scatter(df.x,df.y)
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.scatter(df.x,df.y)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.scatter(df.x,df.y)
plt.plot(df.x,df.y, linewidth=0.2,color='red')
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.show()
```

```python
cov = np.cov(np.array(df).T)
cov
```

```python
eig = np.linalg.eig(cov)
eig_vectors = eig[1]
eig_vectors
# eig_values = eig_values[[i[0] for i in sorted(enumerate(eig_values),reverse=True)]]
# eig_vectors = eig_vectors[[i[0] for i in sorted(enumerate(eig_values),reverse=True)]]
```

```python
new = df @ eig_vectors
new.columns = ['Principal Component 1','Principal Component 2']
new
```

```python
plt.figure(figsize=(20,8))
plt.scatter(new['Principal Component 1'],new['Principal Component 2'])
plt.axhline(y=new['Principal Component 2'][0], linewidth=0.5, color='red')
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.show()
```
